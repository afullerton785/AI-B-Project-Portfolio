As is considered common with AI generation, much of the illustration is not entirely accurate, with the taxi on the pavement and the train on the road in front of a car. Granted, this is not much of an issue for this project as this was more about bringing forth the idea of travel rather than the accuracy of it, however it is worth noting for any future projects.

### Human-AI Collaboration

This project has highlighted how AI models and humans can interpret items of cultural heritage differently. It has found that both sides come with flaws in their interpretations. For AI models this arises through hallucinations, struggling with interpreting limbs, and from protocols altering how the model behaves. For humans these flaws are the inability to show more than one mode of transport at once, and the difficulty of perfectly posing in the same way as the original illustration.

However, both sides have benefits in their own ways, with AI able to produce arguably more detailed images, whereas human recreations copuld be argued to contain a ceratin emotive quality, more of the 'soul' from the original illustrations.

These flaws and benefits come together to form the central argument of this project: humans and AI models can be more effective working creatively together, rather than as separate entities. Instead of pitting both against each other, it would be beneficial to pair them together, in the likelihood of achieving an outcome greater than what either side could achieve alone.
