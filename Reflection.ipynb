### AI Generation Analysis

```python
from IPython.display import display, HTML

img1 = "Travel.jpeg"
img2 = "Gemini_Travel_Page.jpg"

html_str = f"""
<div style="display: flex; justify-content: center;">
    <img src="{img1}" style="width: 45%; margin: 5px;">
    <img src="{img2}" style="width: 45%; margin: 5px;">
</div>
"""

display(HTML(html_str))
```

This generation could be argued to be the one that changed the most. It completely transformed the original illustration from the horse drawn carriage and the bus in the backgrounds to a depiction of a bustling city centre featuring cars, bikes, taxis buses and a train. This does a fantastic job at reflecting the options for travel to football games in the modern day, however does have some flaws. As is considered common with AI generation, much of the illustration is not entirely accurate, with the taxi on the pavement and the train on the road in front of a car. Granted, this is not much of an issue for this project as this was more about bringing forth the idea of travel rather than the accuracy of it, however it is worth noting for any future projects.
