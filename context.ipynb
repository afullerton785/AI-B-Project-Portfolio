# Project Context #


During the trip to the University Lobrary Special Collections I made the decision to centre my project around the piece 'Free Kicks at Football', a football pamphlet from 1882. The pamphlet contains a mixture of cartoon-like illustrations and pieces of writing showing what football was like in that time, both in the game and surrounding it alike. These pieces generally feature an upbeat, humorous tone, reflecting the attitudes of the time to the rising popularity of football. Not much is known about the authors or illustrators of the pamphlet, with some using pseudonyms. 

All information on 'Free Kicks at Football' taken from the University of Glasgow Special Collection Department (2004) 
https://www.gla.ac.uk/myglasgow/library/files/special/exhibns/month/jan2004.html#:~:text=This%20rare%20piece%20of%20football,of%20life%20in%20Victorian%20Glasgow. 

The piece drew my attention straight away as an avid football fan, so I knew prior to the visit what I would be looking for. Once in the special collection, I flicked through the pages one at a time and photographed multiple pages I thought could possibly be interesting for a project. I settled on an idea to use the illustrations in the pamphlet to explore how generative AI and human reinterpretation reshape historical meaning, through AI modernising the illustrations and humans photographing the equivalent.
